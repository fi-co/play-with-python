```python
# create a 2D array of random values between 0 and 1
A = np.random.rand(5,5)
print (A)
```

```python
# compute the mean along both the rows and the columns of A
mean_cols = np.mean(A, axis = 0)  # axis = 0 -> computes the mean along COLUMNS
mean_rows = np.mean(A, axis = 1)  # axis = 1 -> computes the mean along ROWS
print ("Columns means are:", mean_rows)
print ("Rows means are:", mean_cols)
```

```python
# normalize the array to [0,1]

# ------------
#    np.min() Finds the minimum
#    np.max() Finds the maximum
#    normalization: subtract the minimum v from each element and then divide by the range (max - min)
# ------------

norm_A = (A - np.min(A)) / (np.max(A) - (np.min(A)))
print (norm_A)
```

signal processing exercise
you have a noisy EEG signal. the goal is to clean it up using a low-pass filter to remove high-frequency noise and then visualize the signal before and after the filtering.

```python
import pandas as pd
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
# create a toy time-series (EEG recording 1 datapoint every 1 ms)
#   - The sampling rate is the rate at which you record data points.
#   Its usually independent of the oscillation frequency of the signal. -
t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi *10*t) + np.random.randn(len(t))
signal = np.array(signal)
#   " + 0.5 * np.random.randn(len(t)) " add some random noise
```

Signal before filtering:

```python
# plot the signal
plt.plot(t, signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal')
plt.show
```

apply low - filter pass (>30 Hz). \
`scipy.signal.butter()` takes 3 main parameters: the **filter order**(which affects how sharp the cutoff is), the **cutoff frequency** and the **type of filter.**

```python
from scipy.signal import butter, filtfilt
```

```python
# sampling rate
fs = 1000
# cutoff frequency
fc = 30

# design a Butterworth low-pass filter
order = 4
nyquist = 0.3 * fs # nyquist frequency = the highest frequency one can accurately record at given fs
normal_cutoff = fc / nyquist
b, a = butter(order, normal_cutoff, btype='low', analog=False)

# apply the filter to the signal
filtered = filtfilt(b, a, signal)
```

```python
# visualize before and after the filtering
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True, figsize=(10, 6))
ax1.plot(t, signal)
ax1.set_ylabel('Amplitude')
ax1.set_title('Signal')
ax2.plot(t, filtered)
ax2.set_xlabel('Time (s)')
ax2.set_ylabel('Amplitude')
ax2.set_title('Filtered Signal')
```

## Simulating reaction times using Probability Distributions

### Simulate reaction times with a normal distributions

Reaction times in cognitive tasks are often modeled using a norm(), where most people respond near the mean, with fewer responses happening much faster or slower. We can simulate a set of reaction times for a cognitive task using a normal distribution and then analyze and visualize the data.

```python
from scipy.stats import norm, skew, probplot
import numpy as np
import matplotlib.pyplot as plt
```

```python
'''
------------------

step 1: simulate reaction times
let's assume mean_rt = 500 ms and std = 50 ms

------------------
'''
# set parameters for the normal distribution
mean_rt = 500
std_rt = 50
trials = 1000

# simulate reaction times using a norm()
reaction_times = norm.rvs(loc = mean_rt, scale = std_rt, size = trials)
```

```python
# visulize the distribution of rt
plt.hist(reaction_times, bins = 30, density = True)
plt.xlabel ('Reaction Time (ms)')
plt.ylabel ('Probability Density')
plt.title ('Distribution of reaction times')
plt.show()
```

compute descriptive statistics: Mean, std, and skewness. than fit experimental rt to a normal distribution, to check how well data approximate to it.

```python
mean_sim = np.mean(reaction_times)
std_sim = np.std(reaction_times)
skewness = skew(reaction_times)

'''
-------------

  why computing the mean and the std again?
  Sampling introduces randomness: even though we generated data from a norm()
  with known parameters, the actual sample we get will have slitgh variations.

  In real experiments we don't know the true parameters in advance, generally.
  We can only ewstimate them from the data we collect. Computing the mean and
  the std from the sample gives us an estimate of the "true" parameters.


-------------
'''

print ('Mean reaction time:', mean_sim)
print ('Standard deviation:', std_sim)
print ('Skewness:', skewness)
```

```python
# fit a normal distribution
mu, std = norm.fit(reaction_times)

'''
-------------


fitting a norm() to the data mimics the real-world scenario where you
don't know the underlying distribution. In an actual experiment, you
might assume the data follows norm(). By fitting a distribution,
you're testing if that is reasonable.

N.B: in real data analysis, it's good practice to test different distributions.

--------------
'''

#plot a histogram and the fitted distribution
plt.hist(reaction_times, bins = 30, density = True, label = 'Data')

# plot the fitted normal distribution
x = np.linspace(min(reaction_times), max(reaction_times), 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth = 2, label = 'Fitted Normal Distribution')
plt.xlabel ('Reaction Time (ms)')
plt.ylabel ('Probability Density')
plt.title ('Distribution of reaction times')
plt.legend()
plt.show()

#########################

# alternatively, normality can be checked with a Q-Q plot

probplot(reaction_times, dist = 'norm', plot = plt)
plt.show()
```

### Simulate reaction times using a log-normal distribution

The log-normal distribution is often a better fit for reaction times due to their positive skew (most reaction times are fast, but there can be some much slower responses). A log-normal distribution can capture this by modeling the logarithm of the variable as normally distributed.

```python
from scipy.stats import lognorm
```

```python
# parameters of the log-normal ()
shape = 0.2 # control the spread ~ std
scale = np.exp(6) # control the mean
trials = 1000

# simulate reaction times with log-normal distribution
reaction_times = lognorm.rvs(s = shape, scale = scale, size = trials)
```

```python
# plot the histogram from the sample
plt.hist(reaction_times, bins = 30, density = True)
plt.xlabel ('Reaction Time (ms)')
plt.ylabel ('Probability Density')
plt.title ('Distribution of reaction times')
plt.show()
```

```python
# fit a log-normal distribution
shape, loc, scale = lognorm.fit(reaction_times)

# plot a histogram and the fitted distribution
plt.hist(reaction_times, bins = 30, density = True, label = 'Data')

x = np.linspace(min(reaction_times), max(reaction_times), 100)
p = lognorm.pdf(x, shape, loc, scale)
plt.plot(x, p , 'r', linewidth = 2, label = 'Fitted Log-Normal Distribution')
plt.xlabel ('Reaction Time (ms)')
plt.ylabel ('Probability Density')
plt.title ('Distribution of reaction times')
plt.legend()
plt.show()
```

```python
probplot(reaction_times, dist = "lognorm", sparams = shape, plot = plt)
plt.show()
```
